```python
import pandas as pd
import numpy as np
import os
```

```python
job_df = pd.read_csv("../Data/job_description_dataset.csv")
resume_df = pd.read_csv("../Data/resume_dataset.csv")
```

```python
print("Job Dataset:", job_df.shape)
print("Resume Dataset:", resume_df.shape)
```

```python
print(job_df.isnull().sum())
print(resume_df.isnull().sum())
```

```python
job_df["Full_Job_Description"] = (
    job_df["Title"].astype(str) + " " +
    job_df["Skills"].astype(str) + " " +
    job_df["Responsibilities"].astype(str) + " " +
    job_df["Keywords"].astype(str)
)
```
