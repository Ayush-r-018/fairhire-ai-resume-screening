```python
import pandas as pd
import numpy as np

from sentence_transformers import SentenceTransformer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
import os

print(os.getcwd())
```

```python
job_df = pd.read_csv("../Data/job_description_dataset.csv")
resume_df = pd.read_csv("../Data/resume_dataset.csv")

print(job_df.shape)
print(resume_df.shape)
```

```python
print("Job Columns:")
print(job_df.columns.tolist())

print("\nResume Columns:")
print(resume_df.columns.tolist())
```

```python
job_df["Full_Job_Description"] = (
    job_df["Title"].astype(str) + " " +
    job_df["Skills"].astype(str) + " " +
    job_df["Responsibilities"].astype(str) + " " +
    job_df["Keywords"].astype(str)
)
```

```python
print("===== JOB DESCRIPTION SAMPLE =====")
print(job_df["Full_Job_Description"].iloc[0][:1000])

print("\n")
print("===== RESUME SAMPLE =====")
print(resume_df["Resume"].iloc[0][:1000])
```

```python
sample_job = job_df["Full_Job_Description"].iloc[0]

sample_resumes = resume_df["Resume"].head(5)

print("Job Loaded Successfully")
print("Number of Resumes:", len(sample_resumes))
```

```python
from sentence_transformers import SentenceTransformer

model = SentenceTransformer("all-MiniLM-L6-v2")

print("Model Loaded Successfully")
```

```python
job_embedding = model.encode(sample_job)

resume_embeddings = model.encode(
    sample_resumes.tolist()
)

print("Job Embedding Shape:", job_embedding.shape)
print("Number of Resume Embeddings:", len(resume_embeddings))
```

```python
from sklearn.metrics.pairwise import cosine_similarity

scores = []

for i, resume_embedding in enumerate(resume_embeddings):

    similarity = cosine_similarity(
        [job_embedding],
        [resume_embedding]
    )[0][0]

    scores.append(
        {
            "Resume_ID": i + 1,
            "Similarity_Score": similarity
        }
    )

scores
```

```python
scores_df["Match_Percentage"] = (
    scores_df["Similarity_Score"] * 100
).round(2)

scores_df
```

```python
software_resumes = resume_df[
    resume_df["Role"].str.lower() == "software engineer"
]

print("Number of Software Engineer Resumes:")
print(len(software_resumes))

software_resumes.head()
```

```python
job_embedding = model.encode(sample_job)

resume_embeddings = model.encode(
    sample_resumes.tolist()
)
```

```python
scores = []

for i, resume_embedding in enumerate(resume_embeddings):

    similarity = cosine_similarity(
        [job_embedding],
        [resume_embedding]
    )[0][0]

    scores.append(
        {
            "Resume_ID": i + 1,
            "Similarity_Score": similarity
        }
    )

scores_df = pd.DataFrame(scores)

scores_df["Match_Percentage"] = (
    scores_df["Similarity_Score"] * 100
).round(2)

scores_df = scores_df.sort_values(
    by="Similarity_Score",
    ascending=False
)

scores_df
```
